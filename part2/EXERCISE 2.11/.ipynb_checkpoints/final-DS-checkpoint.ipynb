## Helping Functions for Image Processing

```python
# ----- Loading Images ------
def load_rgb_image(image_path):
  """ Loading rgb/color-images """
  image = Image.open(image_path)
  r, g, b = image.convert('RGB').split()
  rgb_image = np.array([np.array(r), np.array(g), np.array(b)])
  return rgb_image

#----- Normalization --------
def normalize_rgb_image(rgb_image):
  """ normalizing images on all color channels """
  normalized_img = []
  for m in rgb_image:
    normalized_img.append((m/255))

  return np.array(normalized_img)

#----- Range return --------
def range_return(channel):
  """ returns channel where each element is in range 0 and 255 """
  channel[channel > 255] = 255
  channel[channel < 0] = 0
  channel = channel.astype(np.uint8) 
  return channel


# ------ Visualization ------

def show_image(rgb_matrix):
  result = Image.merge('RGB', ((Image.fromarray(rgb_matrix[0])), (Image.fromarray(rgb_matrix[1])), (Image.fromarray(rgb_matrix[2]))))
  return result

def image_grid(number_of_rows, number_of_cols,titles, *images):
    """ creates an image grid of specified sizes ( rows x columns ). As an input any number of np.array can be given """
    #applies function for each np.array in a tuple and convert back into tuple
    image_list = []
    for i in images:
        image_list.append(show_image(i))
    images = tuple(image_list)
    
    fig = plt.figure(figsize=(12., 12.)) # specified sizes of each image
    
    # creates grid of specified size, the distance between axes in inch, makes all features shared between all images
    grid = ImageGrid(fig, 111, nrows_ncols=(number_of_rows, number_of_cols), axes_pad=0.4, share_all = True)

    # removes axes
    grid[0].get_yaxis().set_ticks([])
    grid[0].get_xaxis().set_ticks([])

    # puts all images in the grid
    idx = 0
    for ax, im in zip(grid, images):
        ax.set_title(titles[idx], fontdict=None, loc='center', color = "k")
        ax.imshow(im)
        idx +=1

    return plt.show()
```

# Image Processing - Matrix Transformations

In this section we have transform the positioning of elements within matrices and matrix shapes.

### Image rotations

```python
def image_rotation(input_image, direction = "right", times = 1):
    """ rotates an image to both directions specified number of times. 
    Uses matrix_rotate and matrix_back_rotate methods from sMatrix class"""
    
    #In case of colored image
    if len(input_image.shape) > 2:
      rotated = []

      if direction == "right":
        for channel in input_image:
          c = mo.sMatrix(0, 0, 0, channel)
          rotated.append(c.matrix_rotate(times = times))
      
      elif direction == "left":
        for channel in input_image:
          c = mo.sMatrix(0, 0, 0, channel)
          rotated.append(c.matrix_back_rotate(times = times))
    
    #In case of single matrix (gray/binary)
    else:
      if direction == "right":
        img_so = mo.sMatrix(0, 0, 0, input_image)
        rotated = img_so.matrix_rotate(times = times)
        
      elif direction == "left":
        img_so = mo.sMatrix(0, 0, 0, input_image)
        rotated = img_so.matrix_back_rotate(times = times)
    
    
    return np.array(rotated)
```

### Image flipping

```python
def image_mirroring(input_image): #should add for 2d
    """ mirrors an image by using matrix_flip_vertical() method from sMatrix class """
    
    
    if len(input_image.shape) > 2:
      mirrored = []
      for channel in input_image:
        c = mo.sMatrix(0,0,0,channel) #each channel is represented as aseperate matrix which casted to sMatrix class
        mirrored.append(c.matrix_flip_vertical())

    else:
      img_so = mo.sMatrix(0,0,0, input_image)
      mirrored = img_so.matrix_flip_vertical()

    mirrored = np.array(mirrored)    
    return mirrored
```

### Image transition

```python
def image_transition(input_image1, input_image2, t):
    """ creates a matrix as one of the steps of transition from image 1 to image 2, using M(t) = (1-t)*A+t*Z formula. 
    Uses several methods from the sMatrix: matrix_scalar and matrix_addition as well as helper method: normalized_channel"""
    
    if t > 1 or t < 0:
      return "value of t should be between 0 and 1"

    if len(input_image1.shape) > 2:
      transition = []
    
      for channel in range(len(input_image1)):
        #initiates an instance
        c = mo.sMatrix(0,0,0, input_image1[channel])
        d = mo.sMatrix(0,0,0, input_image2[channel])

        #multiples by the scalar
        c = mo.sMatrix(0,0,0, c.matrix_scalar(1-t))
        d = mo.sMatrix(0,0,0, d.matrix_scalar(t))

        #adds to matrices together
        two_matrices = c.matrix_addition(d)

        #two_matrices = (input_image1[i].matrix_scalar(1-t)).matrix_addition(input_image2[i].matrix_scalar(t))
        # makes a range 0 - 255 and changes the type
        transition.append(range_return(two_matrices))
    else:
      img1 = mo.sMatrix(0,0,0, input_image1)
      img2 = mo.sMatrix(0,0,0, input_image2)

      #multiples by the scalar
      img1 = mo.sMatrix(0,0,0, img1.matrix_scalar(1-t))
      img2 = mo.sMatrix(0,0,0, img2.matrix_scalar(t))

      #adds to matrices together
      transition = img1.matrix_addition(img2) 
      
    return transition
```

# Image Processing - Color Manipulation

### Channel amplifier

```python
def color_amplifier(rgb_image, color, scalar = 0.3):
    """ amplifies one of the 3 channels in the matrix: red, green or blue. 
    To add more chosen color scalar should be positive, to remove this color scalar should be negative. 
    Scalar should be in the range of -1 and 1"""
    
    #checks range of input
    if scalar < -1.0 or scalar > 1.0: 
        return "scalar should be between -1.0 and 1.0"

    #defines the constant matrix
    b = 255 * scalar

    constant_matrix = mo.sMatrix(rgb_image[0].shape[0],rgb_image[0].shape[1],b)

    image_adjusted = [] 
    
    # determines which matrix should be modified
    if color == "red":
        leading_channel = 0
    elif color == "green":
        leading_channel = 1
    elif color == "blue":
        leading_channel = 2
    else: 
        print("Color specified incorrectly")
    
    for channel in range(len(rgb_image)):
        c = mo.sMatrix(0,0,0, rgb_image[channel])
        #modifies the main matrix by adding constant
        if channel == leading_channel:
            channel_adjusted = c.matrix_addition(constant_matrix)
          
            image_adjusted.append(range_return(channel_adjusted))
        #appends all other matrices
        else:
            image_adjusted.append(c.values)           
          
    return image_adjusted
```

### Grayscale transformation

```python
def rgb_to_grayscale(rgb_image):
  """ takes an image of three or more channels and computes the average value over the given channel matrices. Returns a two-dimensional numpy array of gray values."""
  
  normalized_image = normalize_rgb_image(rgb_image)
  
  c1 = mo.sMatrix(0,0,0, normalized_image[0])
  c2 = mo.sMatrix(0,0,0, normalized_image[1])

  #computing the sum of elements in the same position over all channel matrices
  summed_matrix = mo.sMatrix(0,0,0, c1.matrix_addition(c2))
  
  if normalized_image.shape[0] > 2:
    for i in range(2, (normalized_image.shape[0])):
      c = mo.sMatrix(0,0,0, normalized_image[i])
      summed_matrix = mo.sMatrix(0,0,0, summed_matrix.matrix_addition(c))
  
  
  #averaging sum over number of channels
  scalar = 255/normalized_image.shape[0]
  grayscale_image = summed_matrix.matrix_scalar(scalar) #averages the channel sums over no. of channels

  return np.uint8(grayscale_image)
```

### Binary transformation

```python
def image_binarize(input_image, threshold = (255/2)):
  """ Creates a binary representation of input image according to threshold (max intesity value/2) 
  by default or user sepcified value in range 0 to 255. Based on gray-scale representation """
  
  if threshold < 0 or threshold > 255: return "threshold value should be between 0 and 255."

  if len(input_image.shape) > 2: 
    input_image = rgb_to_grayscale(input_image) #needs a grayscale input

  c = mo.sMatrix(0,0,0, input_image)
  for m in range(c.n_rows):
    for n in range(c.n_cols):
      if c.values[m,n] > threshold: c.values[m,n] = 255 # replaces values above threshhold to max intesity
      else: c.values[m,n] = 0 # replaces values below threshhold to min intesity

  segmented_img = np.uint8(input_image)
  
  #segmented_img = (input_image > threshold).astype(int) #replaces values with boolean according to values above or below threshold and replace with binary values
  #segmented_img[segmented_img == True] = 255 

  return segmented_img
  
```

### Image brightness

```python
def brighten_image(input_image, scalar=0.3):
  """Adding a positive constant to all of the image pixel values makes the image brighter. 
  Similarly, we can subtract a positive constant from all of the pixel values to make the image darker. """
  #check range of input
  if scalar < -1.0 or scalar > 1.0: 
    return "scalar should be between -1.0 and 1.0"

  #defining the constant matrix
  b = 255 * scalar
  
  # in case of a multidimensional image input
  if len(input_image.shape) > 2: #in case of rgb-image or other multidimensional image inputs
    image_adjusted = []
    constant_matrix = mo.sMatrix(input_image[0].shape[0],input_image[0].shape[1],b)
    
    for channel in input_image:
      c = mo.sMatrix(0,0,0,channel) #each channel is represented as seperate matrix which casted to sMatrix class
        
      channel_adjusted = c.matrix_addition(constant_matrix)
      image_adjusted.append(range_return(channel_adjusted))
      
    return image_adjusted

  # in case of a 2D-image input
  else: 
    c = mo.sMatrix(0,0,0,input_image)
    constant_matrix = mo.sMatrix(input_image.shape[0],input_image.shape[1],b)
    channel_adjusted = c.matrix_addition(constant_matrix) #adding scaled constant
    image_adjusted = range_return(channel_adjusted)
    
  return image_adjusted
```

### Image contrast

```python
def image_contrast(rgb_image, scalar = 1.3):
    """ changes the contrast of the image. To increase the contrast, the scalar should be more than 1.
    To decrease the contrast, scalar should be in the range of 0 and 1. If scalar is 1, the input image will be returned """
    
    #checks range of input
    if scalar < 0: 
        print("scalar should be greater or equal to 0")
    
    else:
        contrasted_matrix = []
        #creates the matrix with the integer to make matrices operations
        constant_matrix = mo.sMatrix(rgb_image[0].shape[0],rgb_image[0].shape[1],128)
        
        for channel in rgb_image:
            c = mo.sMatrix(0,0,0,channel)
            #changes the contrast of the image using ()=(128)+128 formula
            # subtracts 128 from each matrix element
            c.matrix_subtraction(constant_matrix)
            # multiples scalar and each matrix element
            c = mo.sMatrix(0,0,0,c.matrix_scalar(scalar))
            # adds 128 to each matrix element
            new = c.matrix_addition(constant_matrix)
            contrasted_matrix.append(range_return(new))
    
        return contrasted_matrix
```

### Image negative

```python
def image_negative(input_image):
  """ Inverses colors in the image """
  #in case of multiple channels
  if len(input_image.shape) > 2:
    neg_img = []

    for channel in input_image:
      #creates intesity matrix with values 255
      max_intensity_matrix = mo.sMatrix(channel.shape[0],channel.shape[1], 255)
      c = mo.sMatrix(0,0,0,channel)

      # subtracts values from intensity matrix to get inverse image by colors
      max_intensity_matrix.matrix_subtraction(c)
      neg_img.append(max_intensity_matrix.values)

    return np.uint8(np.array(neg_img))
  
  #in case of gray image
  else:
    c = mo.sMatrix(0,0,0,input_image)
    max_intesity_matrix = mo.sMatrix(input_image.shape[0],input_image.shape[1],255)
    neg_img = max_intesity_matrix.matrix_subtraction(c)

  return np.uint8(neg_img)
```

# Image Processing - Quality transformations

### Blurring of the image

```python
def blur_image(input_image, scalar=4):
  " input should be between 0 (most blurry) and 10 (least blurry)"
  

  #in case of rgb or other multidimensions apply blurring on each pixel in each channel
  if len(input_image.shape) > 2:
    blurred_image = []
    for channel in input_image:
      c = mo.sMatrix(0,0,0,channel)
      for m in range(0, c.shape[0]-1):
        for n in range(0, c.shape[1]-1):
          #weighted avg of neighbors (north, south, west, east)
          c.values[m,n] = ((1*scalar)*c.values[m,n] + c.values[m+1,n] + c.values[m-1,n] + c.values[m,n+1] + c.values[m,n-1])/(scalar + 4) 
      blurred_image.append(c.values)
    return np.uint8(blurred_image)

  #in case of single matrix
  else:
    c = mo.sMatrix(0,0,0,input_image)
    for m in range(1, c.n_rows-1):
      for n in range(1, c.n_cols-1):
        c.values[m,n] = ((1*scalar)*c.values[m,n] + c.values[m+1,n] + c.values[m-1,n] + c.values[m,n+1] + c.values[m,n-1])/(1*scalar + 4)
    blurred_image = np.uint8(c.values)
    return blurred_image

    
```

### Images subtraction

```python
def image_difference(input_image1, input_image2):
    """ as input takes two images and find the diference between them by subtracting the second one from the first
    Uses a method from the Smatrix class: matrix_subtraction """
    if len(input_image1.shape) > 2:
      difference = []
      for channel in range(len(input_image1)):       
          c = mo.sMatrix(0,0,0,input_image1[channel].astype(int))
          d = mo.sMatrix(0,0,0,input_image2[channel].astype(int))
          d.matrix_subtraction(c) 
          # normilized and changed type of each channel
          difference.append(range_return(d.values))
    else: 
      img1 = mo.sMatrix(0,0,0,input_image1)
      img2 = mo.sMatrix(0,0,0,input_image2)
      difference = img2.matrix_subtraction(img1)
      difference = range_return(difference)
    return difference
```

### noise reduction

```python
def noise_reduction(input_image):
    """ reduces noize from the image by smoothing values using median filter"""
    if len(input_image.shape) > 2:
      noise_free = []
      for channel in input_image:
        c = mo.sMatrix(0,0,0,channel)
        filtered = c.median_filter()
        noise_free.append(filtered.values)
    else: 
      c = mo.sMatrix(0,0,0,input_image)
      noise_free = c.median_filter()
      noise_free = noise_free.values
    return np.array(noise_free)
```
